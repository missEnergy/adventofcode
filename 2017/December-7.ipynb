```python
import csv

with open("December-7-input.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content] 
content

programs = {}

for row in content:
    name_split = row.split(" (")
    weight_split = name_split[1].split(')')
    if len(weight_split[1]) > 0:
        toppers = weight_split[1][4:].split(", ")
    else:
        toppers = []
    programs[name_split[0]] = {
        'weight': int(weight_split[0]),
        'on_top': toppers
    }
    
programs
```

```python
for program_name in programs.keys():
    flag = True
    for progs in programs.values():
        if program_name in progs['on_top']:
            flag = False
            break;
    if flag:
        print(program_name)
        break;
```

```python
def calc_weight(name):
    if len(programs[name]['on_top']) == 0:
        return programs[name]['weight']
    else:
        return programs[name]['weight'] + sum([calc_weight(x) for x in programs[name]['on_top']])
```

```python
problem_node = 'hlhomy'

root_weights = [calc_weight(top) for top in programs[problem_node]['on_top']]
excessive_weight = max(root_weights) - min(root_weights)

flag = True
while flag:
    leaf_weights = {}
    for top in programs[problem_node]['on_top']:
        leaf_weights[top] = calc_weight(top)
    
    # find the node we want to explore further
    if len(set(leaf_weights.values())) > 1:
        problem_node = max(leaf_weights, key=leaf_weights.get)  
    # we found the problematic node!
    else: 
        flag = False
        
new_weight = programs[problem_node]['weight'] - excessive_weight
new_weight
    
```
