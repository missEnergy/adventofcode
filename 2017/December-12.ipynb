```python
with open("December-12-input.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content] 

connections = []
for row in content:
    connections.append([int(x) for x in row.split(" <-> ")[1].split(", ")])
connections
```

```python
# exercise 1
unique_connectors = set()
unique_connectors.add(0)
flag = True

while flag:
    new_additions = []
    flag = False
    for c in unique_connectors:
        for x in connections[c]:
            if x not in unique_connectors:
                flag = True
                new_additions.append(x)
    for c in new_additions:
        unique_connectors.add(c)

len(unique_connectors)    
```

```python
# exercise 2
unique_connectors = set()
count = 0

for nr in range(len(connections)):
    if nr not in unique_connectors:
        count += 1
        unique_connectors.add(nr)
    flag = True
    while flag:
        new_additions = []
        flag = False
        for c in unique_connectors:
            for x in connections[c]:
                if x not in unique_connectors:
                    flag = True
                    new_additions.append(x)
        for c in new_additions:
            unique_connectors.add(c)
count
```
