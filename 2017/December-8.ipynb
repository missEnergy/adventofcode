```python
import csv

with open("December-8-input.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content] 
content

instructions = []
register_in_bool = []
register_in_instruc = []
for row in content:
    bool_split = row.split(" if ")
    register_in_bool.append(bool_split[1].split(" ")[0])
    instruc_split = bool_split[0].split(" ")
    register_in_instruc.append(instruc_split[0])
    instructions.append({
        'register': instruc_split[0],
        'operation': instruc_split[1],
        'value': int(instruc_split[2]),
        'condition1': bool_split[1].split(" ")[0],
        'condition2': "{} {}".format(bool_split[1].split(" ")[1], bool_split[1].split(" ")[2])
    })
    
results_list = [register_in_bool, register_in_instruc]
results_union = set().union(*results_list)

instructions
```
